```python
# Your code here - remember to use markdown cells for comments as well!
import pandas as pd
import csv
import gzip
basics = pd.read_csv('zippedData/imdb.title.basics.csv.gz', compression ='gzip')
ratings = pd.read_csv('zippedData/imdb.title.ratings.csv.gz', compression ='gzip')
gross = pd.read_csv('zippedData/bom.movie_gross.csv.gz', compression ='gzip')
```

```python
with gzip.open('zippedData/imdb.title.basics.csv.gz', mode='rt', encoding='utf-8') as f:
    df = csv.DictReader(f)
    title = list(df)
```

```python
with gzip.open('zippedData/imdb.title.ratings.csv.gz', mode='rt', encoding='utf-8') as f:
    df1 = csv.DictReader(f)
    ratings = list(df1)
```

```python
with gzip.open('zippedData/bom.movie_gross.csv.gz', mode='rt', encoding='utf-8') as f:
    df = csv.DictReader(f)
    gross = list(df)
```

```python
titles = [i['primary_title'] for i in title]
```

```python
gross1 = []
for j in gross:
    if j['title'] in titles:
        gross1.append(j['title'])
len(gross1)
```
