Firstly, we will check what this datafile contains. This can be done by converting the data into a Pandas DataFrame.

```python
# Checking the size of the dataset
basics.shape
```

```python
# Checking the first five rows
basics.head()
```

This file contains the basic information on the films including their title, start year, runtime_minutes and their respective genres. 

Before performing any descriptive statistics, we need to check if there is any missing data in the dataset. This can be done using the `.isna()` method.

```python
# isna() checks for any missing values and adding .sum() returns the sum of missing values in the columns
basics.isna().sum()
```

The original_title column has 21 missing values, runtime_minutes has 31739 while genres has 5408 values missing. The missing values in the original_title are very few can be replaced with the values in the primary_title without impacting the outcoming of the analysis. This can be done using the `.fillna()` method

```python
# Fill missing values in the 'original_title' column with corresponding values from the 'primary_title' column
basics['original_title'] = basics['original_title'].fillna(basics['primary_title'])
```

One of the objectives of this project is to investigate the ideal runtime. Because of this, the missing values in runtime_minutes column cannot be dropped since this will affect the outcome of the analysis. the best approach will be to fill the missing values with the median of the column. The median is preferred and not the mean since the mean is affected by outliers.

```python
# Fill missing values in the 'runrime_minutes' column with the columns' median
basics['runtime_minutes'] = basics['runtime_minutes'].fillna(basics['runtime_minutes'].median())
```

To deal with the missing data in the genres column, the best approach will be to drop the rows with the missing data since the number of missing values is low and will most likely not affect the outcome of the analysis. `.dropna()` method will be used.

```python
# inplace=True is passed to mutate the dataframe inplace
basics.dropna(inplace=True)
```

Next we'll check if there are any left missing values:

```python
basics.isna().sum()
```

There aren't any left missing values in the dataset. The next step is now getting the descriptive statistics of the dataframe.
